```python
import requests

api_key="DEMO_KEY"
api_url="https://api.nasa.gov"

params={'sol':"1000",'api_key':api_key}
header={'Accept':'Application/json'}

response=requests.get(api_url + "/mars-photos/api/v1/rovers/curiosity/photos/",headers=header,params=params)

print(response.status_code)
print(response.url)


```

```python
data=response.json()
photos_data= data.get('photos')

for photo_data in photos_data:
    print(photo_data['camera']['full_name'])
```

```python
from skimage import io

data=response.json()

photos_data=data.get('photos')

for photo_data in photos_data[4:5]:
    img_url=photo_data['img_src']
    img=io.imread(img_url)
    io.imshow(img)
    print(photo_data['camera']['full_name'])
    
```
